# Global Thresholding Using Otsu's Method 
Implement Otsu's method for binary thresholding of binary images. Say an image has $L$ possible intensity levels. We choose a threshold $t$ to segment the image into two parts. Clearly

$$
0\leq t\leq L-1
$$

Let $F_{i}$ represent the number of times intensity $i$ occurs in the image. Define class 1 ($C_{1}$) as the set of pixels satisfying $\{i : 0\leq i \leq t\}$ and class 2 ($C_{2}$) as the set of pixels satisfying $\{i : t< i \leq L-1\}$. Let the number of pixels in $C_{1}$ and $C_{2}$ be $N_{1}$ and $N_{2}$ respectively. The means and variances of class $j$, $\mu_{j}$ and $\sigma_{j}^{2}$, are defined as

$$
\mu_{j} = \frac{\Sigma_{i\in C_{j}}iF_{i}}{N_{j}}
$$

$$
\sigma^{2}_{j} = \frac{\Sigma_{i \in C_{j}}(i-\mu_{j})^{2}F_{i}}{N_{j}}
$$

Also, the means and variances of the entire image, $\mu_{G}$ and $\sigma^{2}_{G}$, are defined as

$$
\mu_{G} = \frac{\Sigma_{i=0}^{L-1}iF_{i}}{N}
$$

$$
\sigma^{2}_{G} = \frac{\Sigma_{i=0}^{L-1}(i-\mu_{G})^{2}F_{i}}{N}
$$

$N$ is the total number of pixels in the image.

The within class variance, $\sigma^{2}_{w}$, is defined as

$$
\sigma^{2}_{w} = \frac{\Sigma_{j=1}^{2}\sigma_{j}^{2}N_{j}}{N}
$$

The between class variance, $\sigma_{b}^{2}$ is defined as 

$$
\sigma^{2}_{b} = \frac{\Sigma_{j=1}^{2}(\mu_{j}-\mu_{G})^{2}N_{j}}{N}
$$

It is clear that the terms defined above are functions of $t$. $t$ has been skipped for notational brevity. It can be shown that the between class variance can be rewritten as 

$$
\sigma^{2}_{b} = \frac{(\mu_{G}P_{1}-m)^{2}}{P_{1}(1-P_{1})}
$$

$P_{1}$ and $m$ are given as

$$
P_{1} = \frac{N_{1}}{N}
$$

$$
m = \frac{\Sigma_{i\in C_{1}}iF_{i}}{N} = \frac{\mu_{1}N_{1}}{N}
$$

The complete derivation can be found in section 10.3.3 of Digital Image Processing, Third Edition by Rafael C. Gonzalez and Richard E. Woods.
The optimal threshold $t^{*}$ from Otsu's method is given by

$$
t^{*} = argmax_{0\leq t \leq L-1} \sigma^{2}_{b}(t) = argmin_{0\leq t \leq L-1} \sigma^{2}_{w}(t)
$$

It should be noted that there are some slight differences in implementation here compared to what was covered in the class.
- A histogram has not been used. Iterating only over the intensity values present instead of the entire range of possible intensity values is much more efficient and saves redundant computation.
-  The threshold is centered to make it more intuitive. For example, say an image has only two intensity levels $i_{1}, i_{2}$ satisfying $0\leq i_{1} \leq i_{2} \leq L-1$. The code will choose the threshold to $(i_{1}+i_{2})/2$ instead of $i_{1}$ which is more intuitive as it lies at the center of the two intensity levels. Similarly for an image with a clearly bimodal histogram, the chosen threshold will be at the center of the gap between the two sets of intensity values instead of at the end of one of them. The binary image obtained will be the same with either of the thresholds.

```python
def otsu_binary_thresholding(image):
    # convert the image to float
    image = image.astype(np.float)
    # get the sorted set of intensity values in the image 
    thresholds = np.unique(image)
    # get the number of intensity values
    n_intensity_vals = len(thresholds)
    # initialize threshold and best threshold
    threshold = thresholds[0]
    best_threshold = threshold
    # if there is only 1 value, no thresholding is possible
    # if there are 2 intensity values, the threshold must lie in the middle of them
    if n_intensity_vals == 1:
        return 0
    elif n_intensity_vals == 2:
        return np.mean(thresholds)
    # number of pixels in the image
    N = np.size(image)
    # image mean
    mu_t = np.mean(image)
    # initialize pixels in class 1
    N1 = np.sum(image == best_threshold)
    # initialize the cummulative sum
    cumsum = N1*threshold
    # P1 and m as defined above
    m = threshold
    P1 = N1/N
    # initialize sigma_b and best_sigma
    sigma_b = (mu_t*P1-m)**2/(P1*(1-P1))
    best_sigma_b = sigma_b
    for i in range(1, n_intensity_vals-1):
        threshold = thresholds[i]
        Ni = np.sum(image == threshold)
        # update pixels in class 1
        N1 = N1 + Ni
        # update the cummulative sum
        cumsum = cumsum + Ni*threshold
        # recalculate m, P1, and sigma_b
        m = cumsum/N
        P1 = N1/N
        # if sigma_b is higher than the previous best, update best threshold and best_sigma_b
        sigma_b = (mu_t*P1-m)**2/(P1*(1-P1))
        if sigma_b >= best_sigma_b:
            best_sigma_b = sigma_b
            best_threshold = (thresholds[i] + thresholds[i+1])/2
    return best_threshold
```

# Binary Image
Obtain a binary image using Otsu's thresholding method.

```python
def binarized_image(image):
    final_image = np.zeros(np.shape(image), dtype=np.float)
    threshold = otsu_binary_thresholding(image)
    final_image[image >= threshold] = 1
    return final_image
```

# Flatten Image
Flatten an image to apply the k-means algorithm on it. Flattens an image to be of the shape $(x\times y, z)$, where $(x, y)$ are the dimensions of the image. $z$ is $3$ for RGB images and $1$ for grayscale images. 

```python
def flatten_image(image, rgb=True):
    if rgb:
        x, y, _ = np.shape(image)
        flattened_image = np.zeros((x*y, 3))
        for i in range(3):
            flattened_image[:, i] = np.ravel(image[:, :, i])
    else:
        flattened_image = np.ravel(image)
        flattened_image = flattened_image.reshape(-1, 1)
    return flattened_image
```

# K-Means 
Implement the k-means algorithm as discussed in class. The different parameters are described.
- $\textit{data}$ : input data. Can be an image or an array of the shape $(n_{points}, n_{features})$
- $\textit{centroids}$ : initial value of the centroids for the k-means algorithm. If this parameter is not provided, random initializations are chosen and the algorithm is run multiple times. The best clusters formed according to the metric described in the assignment PDF is kept.
- $\textit{k_means}$ : number of centroids
- $\textit{max_iter}$ : maximum number of iterations to run the algorithm
- $\textit{n_inits}$ : when initial value for centroids is not provided, run k-means clustering for $\textit{n_inits}$ number of times and save the best clusters
- $\textit{low}$ : random centroids are chosen with uniform probability from the range ($\textit{low}$, $\textit{high}$)
- $\textit{high}$ : random centroids are chosen with uniform probability from the range($\textit{low}$, $\textit{high}$)
- $\textit{tolerance}$ : stop the algorithm if the average change in the labels between consecutive iterations is less than $\textit{tolerance}$
- $\textit{flattened}$ : boolean parameter which is True if the image is flattened to be of the shape $(n_{points}, z)$. $z$ is $3$ for RGB images and $1$ for grayscale images. If it is false, flatten the image
- $\textit{rgb}$ : boolean parameter which if True if the image is RGB
- $\textit{return_centroids}$ : boolean parameter. Centroids are returned along with the labels if it is True
- $\textit{seed}$ : set a seed for replicability of results. 

```python
def k_means_classifier(data, centroids=None, k_means=3, max_iter=5, n_inits=30,
                       low=0, high=255, tolerance=1e-12, flattened=False, rgb=True,
                       return_centroids=False, seed=None):
    if seed is not None:
        # set a seed
        np.random.seed(seed)
    if not flattened:
        # flatten the data if it is not flattened
        data = flatten_image(data, rgb=rgb)
    n_samples, n_features = np.shape(data)
    # randomly initialize labels
    labels = np.random.randint(low=0, high=k_means, size=n_samples)
    # initialize distances
    distances = np.zeros((k_means, n_samples))
    # for the case when initial points is provided
    if centroids is not None:
        for i in range(max_iter):
            for k in range(k_means):
                # centroids is of the shape (k_means, n_features)
                # centroids[k] is the kth centroid
                c = centroids[k]
                # get the distance of each point from a centroid
                distances[k] = np.sum((data - c)**2, axis=1)
            # get new labels based on distances to centroids calculated
            new_labels = np.argmin(distances, axis=0)
            # check for convergence
            if (np.mean(np.abs(labels-new_labels)) <= tolerance):
                # if convergence has taken place, assign labels and exit
                labels = new_labels
                break
            else:
                # else update labels
                labels = new_labels
                for c in range(k_means):
                    # update centroids
                    centroids[c] = np.mean(data[labels == c], axis=0)
    # when initial points is not provided
    else:
        # initialize variables
        best_cost = np.inf
        centroids = np.zeros((k_means, n_features))
        best_labels = labels.copy()
        for n in range(n_inits):
            # for each run of k-means, randomly initiate centers
            init = np.random.uniform(low, high, size=(k_means, n_features))
            for i in range(max_iter):
                for k in range(k_means):
                    c = init[k]
                    distances[k] = np.sum((data - c)**2, axis=1)
                new_labels = np.argmin(distances, axis=0)
                if (np.min(np.abs(labels-new_labels)) <= tolerance):
                    labels = new_labels
                    break
                else:
                    labels = new_labels
                    avaliable_labels = set(labels)
                    for c in range(k_means):
                        # if a cluster has points, update the corresponding centroid
                        if c in avaliable_labels:
                            init[c] = np.mean(data[labels == c], axis=0)
                        # if there is a cluster with no points, randomly initiate the 
                        # corresponding center again. We could also choose to totally 
                        # ignore this and work with k_means-1 means but we are already 
                        # working with only 3 means and this would lead to pretty bad results
                        else:
                            init[c] = np.random.uniform(low, high, n_features)
            # get the cost as the sum of the minimum distance of each point from some cluster center
            cost = np.sum(np.min(distances), axis=0)
            # update the cost, centroids and labels when you get a lower cost
            if cost < best_cost:
                best_cost = cost
                centroids = init
                best_labels = labels
        labels = best_labels
    # return centroids only if asked for
    if not return_centroids:
        return labels
    else:
        return labels, centroids
```

# Segment Image
Segment an RGB image using the k-means algorithm. The points in a cluster are assigned the colours of their respective centroid. The parameters are the same as those described for the $\textit{k_means_classifier}$ function.

```python
def segment_rgb_image(image, centroids_=None, k_means=3, max_iter=5, n_inits=30, low=0,
                      high=255, tolerance=1e-12, flattened=False, set_centroids=False, seed=None):
    x, y, _ = np.shape(image)
    flattened_image = flatten_image(image, rgb=True)
    final_image = np.zeros_like(flattened_image)
    labels, centroids = k_means_classifier(data=image, centroids=centroids_, k_means=k_means, 
                                           max_iter=max_iter, n_inits=n_inits, low=low, high=high,
                                           tolerance=tolerance, flattened=flattened, rgb=True, 
                                           return_centroids=True, seed=seed)
    # get the labels present
    all_labels = set(labels)
    for label in all_labels:
        if not set_centroids:
            # in this case, to all pixels with label k, assign their average color
            final_image[labels == label] = np.mean(
                flattened_image[labels == label], axis=0)
        else:
            # in this case, to all pixels with label k, assign the kth centroid. The 
            # previous one often gives better results
            final_image[labels == label] = centroids[label]
    final_image = final_image/high
    return final_image.reshape(x, y, 3)
```

# Plotting function
Defining a plotting function for easy plotting of all the cases.

```python
def plotter(image, transformed_image, title_1="", title_2=""):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9,12), constrained_layout=True)
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax1.imshow(image, 'gray')
    ax2.imshow(transformed_image, 'gray')
    ax1.title.set_text(title_1)
    ax2.title.set_text(title_2)
    plt.axis('off')
    plt.show()
```

```python
c_init = np.zeros((3, 3))
c_init[0, 0] = 255
c_init[2] = 255
t1 = otsu_binary_thresholding(palm_leaf_1)
t2 = otsu_binary_thresholding(palm_leaf_2)
print(f"The threshold obtained for 'palmleaf1.pgm' is : {t1}")
print(f"The threshold obtained for 'palmleaf2.pgm' is : {t2}")
```

```python
plotter(palm_leaf_2, binarized_image(palm_leaf_2), "palmleaf2.pgm", "Binary Thresholded Image")
```

Results of image segmentation using the k-means algorithm are presented

```python
plotter(flower, segment_rgb_image(flower, centroids_=c_init, max_iter=5), "flower.png", "Segmented Image")
```

#### Random Initializations
Run k-means for 30 times with random initializations and choose the best clustering. The seeds are fixed after running k-means for a few different values of seeds and choosing the one with the best looking results.

```python
plotter(car, segment_rgb_image(car, max_iter=5, seed=1791278654), "car.ppm", "Segmented Image")
```

```python
plotter(flower, segment_rgb_image(flower, max_iter=5, seed=773890966), "flower.png", "Segmented Image")
```

However, this is a very cherry-picked result. In general, the results are not so appealing and are extremely sentitive to the initial values of centroids. A few results for which seeds have not been carefully set are presented. Since the seeds have not been set, these are suspect to change on different runs of the notebook. First for the car image, and then for the flower image.

```python
plotter(segment_rgb_image(car, max_iter=5), segment_rgb_image(car, max_iter=5))
plotter(segment_rgb_image(car, max_iter=5), segment_rgb_image(car, max_iter=5))
plotter(segment_rgb_image(car, max_iter=5), segment_rgb_image(car, max_iter=5))
plotter(segment_rgb_image(car, max_iter=5), segment_rgb_image(car, max_iter=5))
```

```python
plotter(segment_rgb_image(flower, max_iter=5), segment_rgb_image(flower, max_iter=5))
plotter(segment_rgb_image(flower, max_iter=5), segment_rgb_image(flower, max_iter=5))
plotter(segment_rgb_image(flower, max_iter=5), segment_rgb_image(flower, max_iter=5))
plotter(segment_rgb_image(flower, max_iter=5), segment_rgb_image(flower, max_iter=5))
```

As it is apparent, apart from a few lucky cases, the results are in general not so appealing. Increasing the number of means can make the results much more appealing. However, it does not make sense to use too many clusters as we will end up with a lot clusters which should ideally have been a single cluster. Again, we present results first for the car image and then the flower image.

```python
plotter(segment_rgb_image(car, max_iter=5, k_means=10), segment_rgb_image(car, max_iter=5, k_means=10))
plotter(segment_rgb_image(car, max_iter=5, k_means=10), segment_rgb_image(car, max_iter=5, k_means=10))
plotter(segment_rgb_image(car, max_iter=5, k_means=10), segment_rgb_image(car, max_iter=5, k_means=10))
plotter(segment_rgb_image(car, max_iter=5, k_means=10), segment_rgb_image(car, max_iter=5, k_means=10))
```

As we can see now, the results on an average are much more visually appealing now. However, the bonnet of the car, which should ideally be one cluster, has been assigned different clusters.

```python
plotter(segment_rgb_image(flower, max_iter=5, k_means=5), segment_rgb_image(flower, max_iter=5, k_means=5))
plotter(segment_rgb_image(flower, max_iter=5, k_means=5), segment_rgb_image(flower, max_iter=5, k_means=5))
plotter(segment_rgb_image(flower, max_iter=5, k_means=5), segment_rgb_image(flower, max_iter=5, k_means=5))
plotter(segment_rgb_image(flower, max_iter=5, k_means=5), segment_rgb_image(flower, max_iter=5, k_means=5))
```

We get quite good results for the flower image with much fewer means as there is much less variation in colours and brightness across it compared to the car image.

# Observations
The following observations can be made.
- Otsu's method does a fairly good job at binarizing both the images.
- It does quite a bit better in binarizing $\texttt{palmleaf1.pgm}$ compared to $\texttt{palmleaf2.pgm}$. This can probably be attributed to the fact the illumination in the former is much more uniform.
- Image Segmentation using k-means is extremely sensitive to the initial values chosen for the centroids
- Segmentation results using the given initial values for the centroids on an average look much better than those obtained by running k-means with random iterations multiple times and choosing the best clusters.
- Increasing the number of means to an extent can improve the look of the segmented image. However, it comes with the issue of grouping very similar pixels into different clusters.
