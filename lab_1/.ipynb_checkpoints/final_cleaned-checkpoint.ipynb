```python
# IMPORT
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from skimage.filters import frangi  # ,hessian
from skimage.morphology import convex_hull_image, binary_dilation, skeletonize, medial_axis, thin, area_closing, skeletonize_3d
from scipy.ndimage import binary_erosion
import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import *
%matplotlib inline
```

```python
# Choose an image
choice = 3  # 123
img = None
if(choice == 1):
    img = cv2.imread('cells.png', cv2.IMREAD_UNCHANGED)
if(choice == 2):
    img = cv2.imread(
        'files/gfp_tiles_4_10_2020/dense/dense_with_enhancement.png', cv2.IMREAD_UNCHANGED)
if(choice == 3):
    img = cv2.imread(
        'files/gfp_tiles_4_10_2020/overlapping/overlapping_with_enhancement.png', cv2.IMREAD_UNCHANGED)
img = img[:, :, 1]
og_img = img * 1
```

```python
max_val = int(2**16-1) #16 bit images
threshold1 = 0.2
threhsold2 = 0.2
area_threshold = 54
```

```python
img = img/max_val
img1 = (img > threshold1).astype('uint16')
img2 = (img > threhsold2).astype('uint16')
```

```python
@interact(thresh1=(0, 1, 0.01), thresh2=(0, 1, 0.01))
def show(thresh1, thresh2):
    
    fig, (ax1, ax2) = plt.subplots(1,2,  figsize=(10,5), constrained_layout=True)
    imgx = (img > thresh1).astype('uint16')
    imgy = (img > thresh2).astype('uint16')
    ax1.imshow(imgx, 'gray')
    ax1.set_title(f"Thereshold = {thresh1}")
    ax2.imshow(imgy, 'gray')
    ax2.set_title(f"Thereshold = {thresh2}")
    plt.show()
```

```python
ret, bin_img = cv2.threshold(img2, 0, max_val, cv2.THRESH_BINARY)
bin_img = bin_img.astype(np.uint8)
dt_img = cv2.distanceTransform(bin_img, cv2.DIST_L2, 3)
cv2.normalize(dt_img, dt_img, 0, 1.0, cv2.NORM_MINMAX)

img3 = img * (img2 > 0)
cv2.normalize(img3, img3, 0, 1.0, cv2.NORM_MINMAX)
mdt_img = 0.9*dt_img + 0.1*img3
```

```python
mdt_local_max = fast_local_max(mdt_img)
```

```python
seperated_coordinates, seperated_images = split(
    mdt_local_max, (mdt_local_max > 0))
i_coord = []
j_coord = []
coords = []
mass = []
for i in range(len(seperated_coordinates)):
    x_min, x_max, y_min, y_max = seperated_coordinates[i]
    x, y = centre_of_mass(seperated_images[i])
    i_coord.append(x + x_min)
    j_coord.append(y + y_min)
    coords.append([x + x_min, y + y_min])
    mass.append(np.sum(seperated_images[i]))
coords = merge_coords(i_coord, j_coord, mass, 6)

plt.figure(dpi=100)
plt.imshow(og_img, cmap='gray')
plt.plot(np.array(coords)[:, 1], np.array(coords)[:, 0], 'rx')
plt.axis('off')
plt.title("Maximas")
plt.show()
```

```python
result = frangi(mdt_img, sigmas=[1.1], alpha=1,
                beta=0.9, gamma=0.05, black_ridges=False)
imgf = result > ((np.sum(result)/np.sum(result > 0))*1)
final = bifurcate(imgf)
final2 = ridge_end(imgf)

plt.figure(dpi=100)
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.title("Vessel")
plt.show()
```

```python
bf_coordinates, bf_images = split(final, (final > 0))
bf_i_coord = []
bf_j_coord = []
bf_coords = []
bf_mass = []

for i in range(len(bf_coordinates)):
    x_min, x_max, y_min, y_max = bf_coordinates[i]
    x, y = centre_of_mass(bf_images[i])
    bf_i_coord.append(x + x_min)
    bf_j_coord.append(y + y_min)
    bf_coords.append([x + x_min, y + y_min])
    bf_mass.append(np.sum(bf_images[i]))
bf_coords = merge_coords(bf_i_coord, bf_j_coord, bf_mass, 6)

seperated_coordinates, seperated_images = split(
    mdt_local_max, (mdt_local_max > 0))
i_coord = []
j_coord = []
coords = []
mass = []
for i in range(len(seperated_coordinates)):
    x_min, x_max, y_min, y_max = seperated_coordinates[i]
    x, y = centre_of_mass(seperated_images[i])
    i_coord.append(x + x_min)
    j_coord.append(y + y_min)
    coords.append([x + x_min, y + y_min])
    mass.append(np.sum(seperated_images[i]))
coords = merge_coords(i_coord, j_coord, mass, 6)
```

```python
a = 0.5
net_mass = list(a*np.array(mass))+list((1-a)*np.array(bf_mass))
net_coords = merge_coords(bf_i_coord+i_coord, bf_j_coord+j_coord, net_mass, 6)
```

```python
@interact(show_all=True, thresh=(0, 2, 0.01), sigmain1=(0, 5, 0.01), sigmain2=(0, 5, 0.01),
          nsigma=(1, 15, 1), alphain=(0, 2, 0.01), betain=(0, 2, 0.01), gammain=(1, 15, 0.1),
          a=[0.9999999, 0.999, 0.9, 0.8, 0.7, 0.6, 0.5,
              0.4, 0.3, 0.2, 0.1, 0.001, 0.00000001],
          thin_method=["skeletonize", "skeletonize_3d", "medial_axis", "thin"])
def show(show_all, thresh, sigmain1, sigmain2, nsigma, alphain, betain, gammain, a, thin_method):

    result = frangi(mdt_img, sigmas=list(np.linspace(sigmain1, sigmain2, nsigma)),
                    alpha=alphain, beta=betain, gamma=0.05, black_ridges=False)
    imgf = result > ((np.sum(result)/np.sum(result > 0))*thresh)
    final = bifurcate(imgf, thinning_method=thin_method)
    bf_coordinates, bf_images = split(final, (final > 0))
    bf_i_coord = []
    bf_j_coord = []
    bf_coords = []
    bf_mass = []

    for i in range(len(bf_coordinates)):
        x_min, x_max, y_min, y_max = bf_coordinates[i]
        x, y = centre_of_mass(bf_images[i])
        bf_i_coord.append(x + x_min)
        bf_j_coord.append(y + y_min)
        bf_coords.append([x + x_min, y + y_min])
        bf_mass.append(np.sum(bf_images[i]))
    bf_coords = merge_coords(bf_i_coord, bf_j_coord, bf_mass, 6)
    net_mass = list(a*np.array(mass))+list((1-a)*np.array(bf_mass))
    net_coords = merge_coords(
        bf_i_coord+i_coord, bf_j_coord+j_coord, net_mass, 6)
    
    if show_all:
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(9,9), constrained_layout=True)
        ax1.imshow(og_img, 'gray')
        ax1.plot(np.array(coords)[:,1], np.array(coords)[:,0], 'ro')
        ax1.set_title("mdt maximas")
        ax3.imshow(og_img, 'gray')
        ax3.plot(np.array(bf_coords)[:,1], np.array(bf_coords)[:,0], 'go')
        ax3.set_title("bf points")
        ax2.imshow(skeletonize(imgf), 'gray')
        ax2.plot(np.array(bf_coords)[:,1], np.array(bf_coords)[:,0], 'rx')
        ax2.set_title("Vessel output")
        ax4.imshow(og_img, 'gray')
        ax4.plot(np.array(net_coords)[:,1], np.array(net_coords)[:,0], 'bo')
        ax4.set_title(f"all detections for a={a}")
        plt.show()
    
    else:
        plt.figure(dpi=150)
        plt.imshow(map_of_thinning_methods[thin_method](imgf), 'gray')
        plt.plot(np.array(bf_coords)[:, 1], np.array(bf_coords)[:, 0], 'rx')
        plt.title(f"{thin_method} output")
        plt.show()    
```
